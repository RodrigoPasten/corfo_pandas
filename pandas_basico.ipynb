```python
from faker import Faker
import pandas as pd
import numpy as np
from summarytools import dfSummary
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fake = Faker('es_CL')
nombres = []
for _ in range(1000):
    nombres.append(fake.name())
```

```python
nombres
```

```python
edades = []
for _ in range(1000):
    edades.append(fake.random_int(18,80))
```

```python
edades
```

```python
cedulas_identidad = []
for _ in range(1000):
    cedulas_identidad.append(fake.person_rut())
```

```python
cedulas_identidad
```

```python
altura = []
for _ in range(1000):
    altura.append(fake.pyfloat(1,2,0,1.5,2))
```

```python
altura
```

```python
pesos = []
for _ in range(1000):
    pesos.append(fake.pyfloat(10,2,0,1.5,2))
```

```python
pesos = []
for _ in range(1000):
    pesos.append(fake.random_int(50, 90))
```

```python
pesos
```

```python
df_personas = pd.DataFrame({'Nombre':nombres, 'RUT':cedulas_identidad, 'Edad':edades, 'Altura':altura, 'Peso':pesos})
```

```python
df_personas['Altura']= df_personas['Altura'].round(2)
```

```python
df_personas
```

```python
df_personas.describe().round(3)
```

```python
df_personas['Categoria'] = np.where( df_personas['Edad']<30,'Menor de 30', np.where(df_personas['Edad']>60, 'Mayor de 60', 'Mayor de 30'))
```

```python
df_personas
```

```python

df_personas[['RUT_estandar', 'digito_verificador']] = df_personas['RUT'].str.split('-', expand=True)
df_personas.drop(columns='RUT', inplace=True)
```

```python
df_personas
```

```python
df_personas['IMC'] = (df_personas['Peso'] / df_personas['Altura']).round(3)
```
